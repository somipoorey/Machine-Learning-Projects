## FEATURE SELECTION & DATA SPLIT

###### FEATURE SELECTION

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

X = us_housing.iloc[:,0:-2]  #independent columns
Y = us_housing.iloc[:,-1]    #target column i.e Price

#apply SelectKBest class to extract top 5 best features

bestfeatures = SelectKBest(score_func=chi2, k=5)
fit = bestfeatures.fit(X,Y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  #naming the dataframe columns
print(featureScores.nlargest(5,'Score'))  #print 5 best features
```

###### SPLITTING THE DATA

```python
# Creating Response Variable

Y= us_housing.iloc[:,-2]    #target column i.e Price

Y.head()
```

```python
# Creating Explanatory Variable

X = us_housing.iloc[:,:-2]  #independent columns

X.head()
```

```python
from sklearn.model_selection import train_test_split

rX_train, rX_test, rY_train, rY_test = train_test_split(X,Y, test_size=0.2, random_state=0)
```

## MODELING DATA

###### CREATE A MODEL

```python
# For Linear Regression Model
from sklearn.linear_model import LinearRegression

reg_linreg = LinearRegression(fit_intercept=True,normalize=False,positive=False)
```

```python
# For Lasso Regression Model
from sklearn.linear_model import Lasso

reg_lasso = Lasso(alpha=1,selection='cyclic')
```

```python
# For Decision-Tree Regression Model
from sklearn.tree import DecisionTreeRegressor

reg_decisiontree = DecisionTreeRegressor(criterion='mse',splitter='best')
```

###### TRAIN A MODEL

```python
# For Linear Regression Model
reg_model = reg_linreg.fit(rX_train,rY_train)
```

```python
# For Lasso Regression Model
lasso_model=reg_lasso.fit(rX_train,rY_train)
```

```python
# For Decision-Tree Regression Model
decisiontree_model=reg_decisiontree.fit(rX_train,rY_train)
```

###### EVALUATE A MODEL

```python
# For Linear Regression Model
reg_model.score(rX_train,rY_train)
```

```python
# For Lasso Regression Model
lasso_model.score(rX_train,rY_train)
```

```python
# For Decision-Tree Regression Model
decisiontree_model.score(rX_train,rY_train)
```

```python
coeff_df = pd.DataFrame(reg_model.coef_, X.columns, columns=['Coefficient'])
Intercept_df=pd.DataFrame(reg_model.intercept_, X.columns, columns=['Intercept'])
print(coeff_df)
print('\n',Intercept_df)
```

###### MODEL TUNING

###### Cross Validation

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv= ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

linear_cv=cross_val_score(LinearRegression(),X,Y,cv=cv)
print(linear_cv)
print('\nMean Score:\t',linear_cv.mean())
```

###### Hyperparameter Optimization

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn import metrics

def find_best_model_using_gridsearchcv(X,Y):
    algos = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
                'normalize': [True, False],
                'fit_intercept':[True, False],
                'positive':[True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X,Y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(X,Y)
```

## MODEL EVALUATION & METRICS

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from math import sqrt
```

```python
# Linear Regression Model
MAE = mean_absolute_error(rY_test, linear_price_prediction)
print('Mean Absolute Error: ',MAE)

MSE = mean_squared_error(rY_test, linear_price_prediction)
print('\n')
print('Mean Square Error: ',MSE)

RMSE = sqrt(mean_squared_error(rY_test, linear_price_prediction))
print('\n')
print('Root Mean Squared Error: ',RMSE)

r2 = r2_score(rY_test, linear_price_prediction)
print('\n')
print('r2 Error: ',r2)
```

```python
# Lasso Regression Model
MAE = mean_absolute_error(rY_test, lasso_price_prediction)
print('Mean Absolute Error: ',MAE)

MSE = mean_squared_error(rY_test, lasso_price_prediction)
print('\n')
print('Mean Square Error: ',MSE)

RMSE = sqrt(mean_squared_error(rY_test, lasso_price_prediction))
print('\n')
print('Root Mean Squared Error: ',RMSE)

r2 = r2_score(rY_test, lasso_price_prediction)
print('\n')
print('r2 Error: ',r2)
```

```python
# For Decision-Tree Regression Model
MAE = mean_absolute_error(rY_test, decisiontree_price_prediction)
print('Mean Absolute Error: ',MAE)

MSE = mean_squared_error(rY_test, decisiontree_price_prediction)
print('\n')
print('Mean Square Error: ',MSE)

RMSE = sqrt(mean_squared_error(rY_test, decisiontree_price_prediction))
print('\n')
print('Root Mean Squared Error: ',RMSE)

r2 = r2_score(rY_test, decisiontree_price_prediction)
print('\n')
print('r2 Error: ',r2)
```

## Hypothesis Testing for P-Value Interpretation

```python
from scipy.stats import ttest_1samp
import numpy as np
h_price= us_housing['Price']
h_price_mean = np.mean(us_housing['Price'])
print(h_price_mean)
tset, pval = ttest_1samp(h_price, 0.00)
print('p-values',pval)

if pval < 0.05:     # alpha value of 0.05 or confidence level of 95%
    print(" we are rejecting null hypothesis")
else:
    print("we are accepting null hypothesis")
```
